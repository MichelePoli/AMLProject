```python
import matplotlib.pyplot as plt
import numpy as np

# Dati (estratti dal log fornito)
epochs = np.arange(1, 21)
train_loss_B4 = [
    1.0905, 0.8850, 0.7902, 0.7519, 0.6948, 0.6817, 0.6324, 0.5768, 0.5470,
    0.4887, 0.4741, 0.4629, 0.4479, 0.4334, 0.4181, 0.4090, 0.5618, 0.4131,
    0.4340, 0.4072
]
miou_B4 = [
    0.2734, 0.2365, 0.3045, 0.3081, 0.3014, 0.2719, 0.2857, 0.3155, 0.3267,
    0.3126, 0.2926, 0.3138, 0.3342, 0.3087, 0.3281, 0.3127, 0.3040, 0.3159,
    0.3245, 0.3204
]
final_iou_B4 = [0.45379692, 0.31175447, 0.08664531, 0.4612361, 0.10841636, 0.25179467, 0.5689296]

# Grafico della perdita
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss_B4, label='Train Loss B4', color='blue', marker='o')
plt.title('Training Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Grafico mIoU
plt.subplot(1, 2, 2)
plt.plot(epochs, miou_B4, label='Validation mIoU B4', color='green', marker='o')
plt.title('Mean IoU over Epochs')
plt.xlabel('Epochs')
plt.ylabel('mIoU')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Grafico IoU per classe
classes = ['Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5', 'Class 6', 'Class 7']
plt.figure(figsize=(10, 5))
plt.bar(classes, final_iou_B4, color='orange', alpha=0.7)
plt.title('IoU per Class at Final Epoch B4')
plt.ylabel('IoU')
plt.xlabel('Classes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
