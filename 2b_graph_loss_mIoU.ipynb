```python
import matplotlib.pyplot as plt

# Dati delle epoche
epochs = list(range(1, 21))
training_loss = [1.5659, 1.4198, 1.3409, 1.3190, 1.2453, 1.1687, 1.0970, 1.0742, 1.0111, 0.9755, 
                 0.9730, 0.9772, 0.9064, 0.8961, 0.8637, 0.8653, 0.8591, 0.8514, 0.8285, 0.8657]
validation_loss = [1.7466, 1.6608, 3.4418, 1.9794, 1.4210, 2.8305, 1.3879, 1.5068, 1.4330, 2.1426,
                   1.9133, 7.7969, 1.3797, 1.4108, 1.3927, 1.4053, 1.4149, 1.5898, 1.2113, 1.4527]
validation_miou = [0.1477, 0.1803, 0.1556, 0.1961, 0.2425, 0.1929, 0.2437, 0.2357, 0.2407, 0.2382, 
                   0.2460, 0.2064, 0.3032, 0.2992, 0.2938, 0.2880, 0.2987, 0.2985, 0.3287, 0.2665]

# Creazione del grafico per la perdita (loss)
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(epochs, training_loss, label='Training Loss', marker='o')
plt.plot(epochs, validation_loss, label='Validation Loss', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)

# Creazione del grafico per la mIoU
plt.subplot(1, 2, 2)
plt.plot(epochs, validation_miou, label='Validation mIoU', color='orange', marker='o')
plt.xlabel('Epochs')
plt.ylabel('mIoU')
plt.title('Validation mIoU')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
